```python
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

import numpy as np
import pandas as pd
import plotly.express as px

# Generate 3D scatter plot data
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df_3d = pd.DataFrame(data)

# Create a 3D scatter plot
fig_3d = px.scatter_3d(df_3d, x='X', y='Y', z='Z', title='3D Scatter Plot', color='Z', color_continuous_scale='Viridis')
fig_3d.show()
```

```python
# 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade
# categories.

# Generate violin plot data
np.random.seed(15)
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Create a violin plot
fig_violin = px.violin(df_grades, y='Score', x='Grade', box=True, points='all', title='Distribution of Scores Across Grade Categories')
fig_violin.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.

# Generate heatmap data
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Create a pivot table for the heatmap
heatmap_data = df_sales.groupby(['Month', 'Day'])['Sales'].sum().reset_index()

# Create a heatmap
fig_heatmap = px.density_heatmap(heatmap_data, x='Day', y='Month', z='Sales', 
                                  title='Sales Variation Across Months and Days',
                                  color_continuous_scale='Viridis')

fig_heatmap.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generating x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Computing z values
z = np.sin(np.sqrt(x**2 + y**2))

# Plotting 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
plt.show()
```

```python
# 5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.



np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)


plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5)
plt.title('Bubble Chart of Population vs GDP')
plt.xlabel('GDP')
plt.ylabel('Population')

# Annotating countries
for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=12)

plt.show()
```
