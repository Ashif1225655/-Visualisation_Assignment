```python
# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
# dataset.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(0)  
x = np.random.rand(100) * 100  # 100 random values for variable x (between 0 and 100)
y = 3.5 * x + np.random.randn(100) * 30  # Linear relationship with some noise


data = pd.DataFrame({'Variable X': x, 'Variable Y': y})


plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Variable X', y='Variable Y', color='blue', alpha=0.6, edgecolor='k')

# Add labels and titleplt.title('Scatter Plot of Synthetic Dataset', fontsize=15)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)

# Show the plot
plt.show()



```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories
# based on numerical values.


# Generate a synthetic dataset
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 100  # 100 random values for variable x (between 0 and 100)
y = 3.5 * x + np.random.randn(100) * 30  # Linear relationship with some noise

# Create a DataFrame for better handling with Seaborn
data = pd.DataFrame({'Variable X': x, 'Variable Y': y})

# Create the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Variable X', y='Variable Y', color='blue', alpha=0.6, edgecolor='k')

# Add labels and title
plt.title('Scatter Plot of Synthetic Dataset', fontsize=15)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)

# Show the plot
plt.show()



```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
# variable across different categories.


np.random.seed(0) 
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = {
    'Category A': np.random.normal(loc=20, scale=5, size=100),  # Mean 20
    'Category B': np.random.normal(loc=30, scale=5, size=100),  # Mean 30
    'Category C': np.random.normal(loc=25, scale=5, size=100),  # Mean 25
    'Category D': np.random.normal(loc=35, scale=5, size=100),  # Mean 35
}


data = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate(list(values.values()))
})


plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=data, palette='Set2')


plt.title('Distribution of Values Across Categories (Box Plot)', fontsize=15)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)


plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.


np.random.seed(0)  
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = 2 * x1 + np.random.normal(0, 0.1, n_samples)  # Correlated with x1
x3 = -x1 + np.random.normal(0, 0.1, n_samples)  # Inversely correlated with x1
x4 = np.random.rand(n_samples)  # Independent feature


data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})


correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=15)

# Show the plot
plt.show()
```
