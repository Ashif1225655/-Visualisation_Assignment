```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Data for sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure
p = figure(title="Sine Wave", x_axis_label="X", y_axis_label="Y")

# Add a line renderer
p.line(x, y, legend_label="sin(x)", line_width=2)

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np

output_notebook()

# Random data
np.random.seed(25)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(10, 100, size=100)
colors = np.random.randint(0, 256, size=100)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create scatter plot
p = figure(title="Scatter Plot with Random Data", x_axis_label="X", y_axis_label="Y")
mapper = linear_cmap(field_name="colors", palette=Viridis256, low=min(colors), high=max(colors))
p.scatter('x', 'y', size='sizes', color=mapper, source=source, fill_alpha=0.6)

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Random data for histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(title="Histogram of Data", x_axis_label="Value", y_axis_label="Frequency")

# Plot histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

show(p)

```

```python
#5. Create a Bokeh heatmap using the provided dataset.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np

output_notebook()

# Data for the heatmap
data_heatmap = np.random.rand(10, 10)  # Random 10x10 data
x = np.linspace(0, 1, 10)  # x values
y = np.linspace(0, 1, 10)  # y values
xx, yy = np.meshgrid(x, y)  # Create a meshgrid

# Flatten the data for plotting
data = {
    'X': xx.flatten(),
    'Y': yy.flatten(),
    'Z': data_heatmap.flatten()
}

# Create the plot with correct arguments
p = figure(title="Heatmap of Random Data", x_axis_label="X", y_axis_label="Y", 
           width=400, height=400, toolbar_location=None)

# Define the color mapper for the heatmap
mapper = LinearColorMapper(palette=Viridis256, low=data['Z'].min(), high=data['Z'].max())

# Plot the heatmap
p.rect(x='X', y='Y', width=0.1, height=0.1, source=data, 
       fill_color={'field': 'Z', 'transform': mapper}, line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0), title="Intensity")
p.add_layout(color_bar, 'right')

show(p)
```
